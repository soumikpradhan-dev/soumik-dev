```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset

data = pd.read_csv('drive/MyDrive/new folder/ipldata.csv')

# Drop the 'umpire3' column
data_cleaned = data.drop(columns=['umpire3'])

# Handle missing values in 'city', 'winner', and 'player_of_match'
data_cleaned['city'].fillna('Unknown', inplace=True)
data_cleaned.dropna(subset=['winner', 'player_of_match'], inplace=True)

# Basic Data Analysis
# Number of matches played in each season
matches_per_season = data_cleaned['season'].value_counts().sort_index()

# Number of matches won by each team
matches_won_by_team = data_cleaned['winner'].value_counts()

# Top players by the number of 'Player of the Match' awards
top_players = data_cleaned['player_of_match'].value_counts().head(10)

# Visualizations
# Number of matches played each season
plt.figure(figsize=(12, 6))
sns.barplot(x=matches_per_season.index, y=matches_per_season.values, palette='viridis')
plt.title('Number of Matches Played Each Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()

# Number of matches won by each team
plt.figure(figsize=(12, 6))
sns.barplot(y=matches_won_by_team.head(10).index, x=matches_won_by_team.head(10).values, palette='coolwarm')
plt.title('Top 10 Teams by Number of Matches Won')
plt.xlabel('Number of Matches Won')
plt.ylabel('Team')
plt.show()

# Top players by the number of 'Player of the Match' awards
plt.figure(figsize=(12, 6))
sns.barplot(y=top_players.index, x=top_players.values, palette='inferno')
plt.title('Top 10 Players by Player of the Match Awards')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.show()
```
