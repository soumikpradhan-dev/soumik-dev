```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import pandas as pd

# Convert 'Attendance' and 'RoundID' to numeric if necessary
dt['Attendance'] = pd.to_numeric(dt['Attendance'], errors='coerce')
dt['RoundID'] = pd.to_numeric(dt['RoundID'], errors='coerce')

# Handle missing values (if any) - *Key Change*
# Drop rows where either 'Attendance' or 'RoundID' has a missing value
dt_clean = dt.dropna(subset=['Attendance', 'RoundID'])

# Extract the cleaned data for correlation calculation
x = dt_clean['Attendance']
y = dt_clean['RoundID']

# Calculate correlation coefficients
r1, p1 = scipy.stats.spearmanr(x, y)
print("Pearson correlation:", r1, "p-value:", p1)

r2, p2 = scipy.stats.spearmanr(x, y)
print("Spearman correlation:", r2, "p-value:", p2)

r3,p3 = scipy.stats.kendalltau(x, y)
print("Kendall Tau correlation:", r3, "p-value:", p3)
```
